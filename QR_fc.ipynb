```python
import scipy
import scipy.linalg   # SciPy Linear Algebra Library
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
def scaling(df):
    from sklearn.preprocessing import MinMaxScaler
    scaler=MinMaxScaler(feature_range=(0, 1)).fit(df) 
    scaled_features=scaler.transform(df) 
    return scaled_features
```

## QR Decomposition

```python
Q, R = np.linalg.qr(a_matrix, mode='reduced')
```

```python
Q
```

```python
Q.shape
```

```python
R
```

```python
R.shape
```

```python
R_inv = np.linalg.inv(R)
```

```python
Q_trans = np.transpose(Q)
```

```python
x1 = np.matmul(R_inv, Q_trans)
```

```python
ones = np.ones(1, dtype = None, order = 'C')
```

```python
#B1[2:3]
```

## Calculation of source apportionment

```python
X_arr=[]
x_df = []
for i in range(1,(len(B1)+1)):
    B_arr = B1[(i-1):i]
    #print(B_arr)
    b1=np.asarray(B_arr)
    #print(b1)
    b1=b1.T
    b1=scaling(b1)
    b_matrix = np.vstack([b1,ones])
    X = np.matmul(x1,b_matrix)
    x_df.append(X)
    print(x_df)
    X_sum = sum(X)
    X_arr.append(X_sum)
    #print(X_arr)
    #B_arr=[]
    #print(B_arr)
```
